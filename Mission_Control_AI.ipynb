```python
def analisar_missao(temperatura, energia, comunicacao, status_modulo):

    if temperatura > 80 and energia < 20:
        return """
ALERTA CRÍTICO

Temperatura acima do limite seguro e energia em nível crítico.

Ações recomendadas:
- Ativar resfriamento de emergência.
- Entrar em modo de economia de energia.
- Priorizar sistemas essenciais.
"""

    if temperatura > 80:
        return """
ALERTA DE TEMPERATURA

Temperatura acima do limite recomendado.

Ações:
- Verificar sistema térmico.
- Monitorar sensores.
"""

    if energia < 20:
        return """
ALERTA DE ENERGIA

Nível de bateria crítico.

Ações:
- Ativar modo de economia.
- Reduzir consumo de sistemas secundários.
"""

    if comunicacao == "Instável":
        return """
ALERTA DE COMUNICAÇÃO

Foram detectadas falhas de comunicação.

Ações:
- Reiniciar antenas.
- Executar protocolo de reconexão.
"""

    return """
MISSÃO OPERANDO NORMALMENTE

Todos os parâmetros estão dentro dos limites esperados.
"""
```

```python
temperatura = 30
energia = 85
comunicacao = "Estável"
status_modulo = "Operacional"

analise = analisar_missao(
    temperatura,
    energia,
    comunicacao,
    status_modulo
)

print("=" * 40)
print("MISSION CONTROL AI")
print("=" * 40)
print(f"Temperatura: {temperatura}°C")
print(f"Energia: {energia}%")
print(f"Comunicação: {comunicacao}")
print(f"Status: {status_modulo}")

print("\nANÁLISE:")
print(analise)
```
